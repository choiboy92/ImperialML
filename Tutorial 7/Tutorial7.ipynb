```python
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots() 
X = np.random.normal(loc=0, scale=0.1, size=[40,1])

from sklearn.neighbors import KernelDensity
bandwith = [0.2, 1, 2, 4]

for i in range(0, len(bandwith)):
  kde = KernelDensity(kernel='tophat', bandwidth= bandwith[i]).fit(X)
  x_sample = np.linspace(-5, 5, 1000)
  p = np.exp(kde.score_samples(x_sample.reshape(len(x_sample), -1)))
  ax.plot(x_sample, p, label = bandwith[i])
  ax.legend()
plt.show()
```
